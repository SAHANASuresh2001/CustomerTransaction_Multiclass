```python
# Excluding/separating a categorical column from numerical columns

num_df=merged_df.select_dtypes(exclude=['object','datetime64[ns]'])
```

```python
# Applying a box-plot to check the outliers
for column in num_df:
    plt.figure(figsize=(10,1))
    sns.boxplot(data=num_df,x=column)
    
```

```python
# outlier treatment

def cap_data(num_df):
    for col in num_df.columns:
        print("capping the",col)
        if (((num_df[col].dtype)=='float64')|((num_df[col].dtype)=='int64')):
            percentiles = num_df[col].quantile([0.20,0.80]).values
            num_df[col][num_df[col] <= percentiles[0]] = percentiles[0]
            num_df[col][num_df[col] >= percentiles[1]] = percentiles[1]
        else:
            num_df[col]=num_df[col]
    return num_df
final_num_df = cap_data(num_df)

```

```python
for column in final_num_df:
    plt.figure(figsize=(10,1))
    sns.boxplot(data=final_num_df,x=column)
```

```python
# converting the object data type into categorical data type
cat_df = cat_df.astype("category")
```

```python
# Converting non numerical column into numerical

from sklearn.preprocessing import LabelEncoder

Label = LabelEncoder() 
cat_df['CustomerID'] = Label.fit_transform(cat_df[ 'CustomerID'])
cat_df['DateOfemail'] = Label.fit_transform(cat_df[ 'DateOfemail'])
cat_df['timestamp_x'] = Label.fit_transform(cat_df[ 'timestamp_x'])
cat_df['DateOfOrder'] = Label.fit_transform(cat_df[ 'DateOfOrder'])

cat_df['timestamp_y'] = Label.fit_transform(cat_df[ 'timestamp_y'])
cat_df['DateOfRegistration'] = Label.fit_transform(cat_df[ 'DateOfRegistration'])
cat_df['RetentionPotential'] = Label.fit_transform(cat_df[ 'RetentionPotential'])


```

```python
# coverting the some categorical features into numerical features using OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

df = pd.DataFrame(cat_df)

# Initialize the OneHotEncoder object and specify columns to be encoded
ohe = OneHotEncoder(sparse=False, categories='auto')
columns_to_encode = ['City', 'PreferredDeliveryDay','EmailType','MailOpened','MailClicked']

# Fit and transform the OneHotEncoder on the selected columns of the dataframe
encoded_cols = ohe.fit_transform(df[columns_to_encode])

# Create a dataframe from the encoded columns with the original column names
encoded_df = pd.DataFrame(encoded_cols, columns=ohe.get_feature_names_out(columns_to_encode))

# Combine the original dataframe with the encoded dataframe
cat_df = pd.concat([df.drop(columns_to_encode, axis=1), encoded_df], axis=1)

print(cat_df.head())
```

```python
# concatinating numerical and catrgorical columns 
data_frame=pd.concat([final_num_df,cat_df],axis=1)
```

```python
# Define target variable and list of features
target_var = "RetentionPotential"
features = data_frame.select_dtypes(include='number').columns.tolist()
features.remove(target_var)

# Set correlation threshold
corr_threshold = 0.01

# Drop features with low correlation
for feature in features:
    corr = data_frame[feature].corr(data_frame[target_var])
    if abs(corr) < corr_threshold:
        data_frame = data_frame.drop(columns=[feature])
        print(f"Dropped column {feature} due to low correlation with target variable.")
```

```python
data_frame.shape
```

```python
# separating input columns from output column
X = data_frame.loc[:,data_frame.columns!="RetentionPotential"]
```

```python
y=data_frame["RetentionPotential"]
```

# Train-test split

```python
# importing train_test_split library 
from sklearn.model_selection import train_test_split
```

```python
# Train-test split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=100)
```

 # Feature Scaling

```python
# importing StandardScaler library        
from sklearn.preprocessing import StandardScaler
```

```python
# Applying the standardscaler to training data

sc = StandardScaler().fit(X_train)
```

```python
# Transforming the training data into standard

X_train_std = sc.transform(X_train)
```

# Model Building

```python
#importing LogisticRegression package

from sklearn.linear_model import LogisticRegression

# Create a logistic regression model
log_reg = LogisticRegression(multi_class='multinomial', solver='lbfgs')

# Fit the model to the training data
log_reg.fit(X_train_std, y_train)

# Make predictions on the test data
y_pred = log_reg.predict(X_test)
```

```python
# Importing accuracy-score

from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
# Training the Random Forest Classification model on the Training set

from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 2)
classifier.fit(X_train, y_train)
```

```python
# predicting  the output of test dat
predictions=classifier.predict(X_test)
```

```python
# importing accuracy_score library
from sklearn.metrics import accuracy_score
```

```python
# checking the accuracy score
accuracy_score(y_test,predictions)
```
